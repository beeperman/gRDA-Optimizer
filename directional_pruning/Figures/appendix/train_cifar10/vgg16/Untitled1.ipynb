```python
import re
import numpy as np

def read_file(mu,c,seed):
    name = 'mu_'+str(mu)+'_c_'+str(c)+'_lr_'+str(seed)+'.txt'
    a = open(name,'r')
    tmp = a.readline()
    tmp = re.sub(' +',' ',tmp)
    while len(tmp) != 8:
        tmp = a.readline()
        tmp = re.sub(' +',' ',tmp).split(' ')
        tmp[-1] = tmp[-1].split('\n')[0]
    result = []
#     print(tmp)
    while len(tmp) > 1:
        try:
            tmp = [float(i)  for i in tmp if i is not '']
            result.append(tmp)
        except:
            pass
        tmp = a.readline()
        tmp = re.sub(' +',' ',tmp).split(' ')
        tmp[-1] = tmp[-1].split('\n')[0]
#         print(tmp)
#     print(result)
    return result


record = []

result = []
for seed in [1,2,3,5,6,7,8,9,10]:
    result.append(read_file(0.51,0.0,seed))
record.append(result)

for mu in [0.4,0.51,0.6,0.7]:
    result = []
    for seed in [1,2,3,5,6,7,8,9,10]:
        print([mu,seed])
        result.append(read_file(mu,0.0005,seed))
    record.append(result)




result = np.array(record)

# result
```
