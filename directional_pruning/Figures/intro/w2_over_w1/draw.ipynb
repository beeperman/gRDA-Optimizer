```python
import re
import numpy as np

def read_file():
    name = 'four.txt'
    a = open(name,'r')
    tmp = a.readline()
    result1 = []
    result2 = []
    while tmp:
        tmp = tmp.split('\n')[0].split('\t')
        result1.append([float(tmp[0]),float(tmp[1])])
        try:
            result2.append([float(tmp[2]),float(tmp[3])])
        except:
            pass
        tmp = a.readline()
    return [np.array(result1),np.array(result2)]


result = read_file()
result1 = result[0]
result2 = result[1]
```

```python
import matplotlib.pyplot as plt
import numpy
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

fig, axs = plt.subplots(1, 1)
fig.set_figheight(4)
fig.set_figwidth(6)

axs_index = 0
axs_sub = axs
window_size = 500

linewidth = 2

epoch_num = numpy.array(range(0,200))/2
epoch_num600 = numpy.array(range(0,600))/6

vgg_param = np.sqrt(numpy.ones(600) / 15245130)
color = next(axs_sub._get_lines.prop_cycler)['color']
l1, = axs_sub.plot(epoch_num600, vgg_param, ':', color=color, alpha=0.6, linewidth=linewidth)

l3, = axs_sub.plot(epoch_num600, result1[:600,0], color=color,linewidth=linewidth)

l4, = axs_sub.plot(epoch_num600, result1[:600,1], '--', color=color,linewidth=linewidth)

wrn_param = np.sqrt(numpy.ones(200) / 36546980)
color = next(axs_sub._get_lines.prop_cycler)['color']
l2, = axs_sub.plot(epoch_num, wrn_param, ':', color=color, alpha=0.6, linewidth=linewidth)

l5, = axs_sub.plot(epoch_num, result2[:200,0], color=color,linewidth=linewidth)

l6, = axs_sub.plot(epoch_num, result2[:200,1], '--', color=color,linewidth=linewidth)

axs_sub.set_ylim(0,0.0020)
axs_sub.set_ylabel('$||w||_2 / ||w||_1$', fontsize=18)
axs_sub.set_xlabel('epochs completed / total (%)', fontsize=18)
axs_sub.grid(True)



l7, = axs_sub.plot([0], marker='None',
           linestyle='None', label='dummy-tophead')


axs_sub.grid(True)

plt.rc('legend', fontsize=13)    # legend fontsize
plt.rc('xtick', labelsize=14)    # fontsize of the tick labels
plt.rc('ytick', labelsize=14)    # fontsize of the tick labels

plt.rcParams.update({'font.size': 13})
axs.legend([l3, l4, l1, l5, l6, l2],["gRDA        ", "SGD        ", "$d^{-1/2}$        ", "gRDA  ", "SGD  ", "$d^{-1/2}$  "], loc='upper left', ncol=2, title="CIFAR10-VGG16   CIFAR100-WRN28x10")

plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
plt.savefig("w2_over_w1.pdf", bbox_inches='tight')
```
